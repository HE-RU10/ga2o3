```python
def deduct_baseline(a,b,degree):
# Fit a polynomial to the data
    x=np.zeros(2)
    x[0]=a.iloc[0]
    x[1]=a.iloc[-1]
    y=np.zeros(2)
    y[0]=b.iloc[0]
    y[1]=b.iloc[-1]
    coeffs = np.polyfit(x, y, degree)
    baseline = np.polyval(coeffs, a)
    
# Subtract the baseline from the original signal
    corrected_signal = b - baseline

    return b 
```

```python
gamma=deduct_baseline(Gardf_gamma.iloc[230:270,0],Gardf_gamma.iloc[230:270,1],1)
alpha=deduct_baseline(Gardf.iloc[230:270,0],Gardf.iloc[230:270,1],1)
alpha_ann=deduct_baseline(Gardf_anneal.iloc[230:270,0],Gardf_anneal.iloc[230:270,1],1)
```
