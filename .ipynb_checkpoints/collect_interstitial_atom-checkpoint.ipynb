```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
%matplotlib inline
import seaborn as sns
```

```python
path='/home/heruhe/Desktop/Ga2o3/cascade/5type/1500ev'
os.chdir(path)
#list success cascades info according the obtained data last frame
success_s=[]
for i in range(180):
    if os.path.isfile('./data.lastframe{}'.format(i)):
        success_s.append(i)
success_s=np.array(success_s)
```

```python
#seperate frames into 2 group according to the PKA atom type atom is Ga or O 
pkatype=np.zeros(len(success_s))
for i in range(len(success_s)):
    file="PKA_inf{}.txt".format(success_s[i])
    f=open(file)
    pkatype[i]=int(f.readlines()[0].split()[2])

#frame array which PKA atom is Ga
PKA_Ga_frame=success_s[np.where(pkatype<=2)[0]]
#frame array which PKA atom is O
PKA_O_frame=success_s[np.where(pkatype>2)[0]]
```

```python
#reading all atoms final position
df=pd.read_csv('data.lastframe0', skiprows = 17, nrows=160000,sep=" ")

df=df.reset_index()
df.drop(columns=df.columns[-3:], axis=1, inplace=True)
df.columns=['Particle Identifier','Particle Type','x','y','z']
df
```

```python

#read defects sites original position
odf=pd.read_csv('Idefects_original_position0.csv')
#read interstitial atoms final position
ddf=pd.read_csv('Idefects_displaced_position0.csv')
ddf.drop(columns=ddf.columns[0], axis=1, inplace=True) #delete unneccessary columns
#delete original atoms in the defects sites, obtain interstitial atoms position info
I_atomsdf=ddf.drop(pd.Index(np.where(ddf['Particle Identifier'].isin(odf['Particle Identifier']))[0]))
I_atomsdf.reset_index(drop=True, inplace=True)
I_atomsdf
```

```python
i=89785
#obtain interstitial atom's positon
x=I_atomsdf[I_atomsdf['Particle Identifier']==i].at[0,'x']
y=I_atomsdf[I_atomsdf['Particle Identifier']==i].at[0,'y']
z=I_atomsdf[I_atomsdf['Particle Identifier']==i].at[0,'z']
#regard the interstitial atom as center, choose a cube distrcition, extract all atoms info in the cube  
cube_df = df[(np.abs(df.x-x)<3)& (np.abs(df.y-y)<3)&(np.abs(df.z-z)<3)]
print(df[df['Particle Identifier']==i])
#In the cube, calculate each atom's distance with the center interstitial atom
cube_df.loc[:,'distance']=np.sqrt((cube_df.x-x)**2+(cube_df.y-y)**2+(cube_df.z-z)**2)
cube_df.sort_values(by=['distance'],inplace=True)
cube_df
```
