```python

df=pd.read_csv('occupancy_analysis.csv')
df['sum_occupancy']=df['occupancy.1']+df['occupancy.2']+df['occupancy.3']+df['occupancy.4']+df['occupancy.5']
df
```

```python
#extract intervals data
interval=df[df['sum_occupancy']>1]
o=[]
#In interstitials sites, calculate each type atom's occupancy amounts
n=np.zeros(5)
for i in range(5):
    o.append(interval[interval['occupancy.{}'.format(i+1)]>=1])
    n[i]=o[i].size
#In interstitials sites, calculate each type atom's occupancy percentage
interval_occu=n/sum(n)
interval_occu
```

```python
pka_interval_sites=[]
pka_vacancy_sites=[]
for i in range(5):
    df=pd.read_csv('interstitial_analysis{}.csv'.format(i+1))
    frames=df['frame']
    inv_sites=np.zeros((len(frames),5))
    v_sites=np.zeros((len(frames),5))
    n=0
    for frame in frames:
        
        odf=pd.read_csv('occupancy_analysis{}.csv'.format(frame))
        odf['sum_occupancy']=odf['occupancy.1']+odf['occupancy.2']+odf['occupancy.3']+odf['occupancy.4']+odf['occupancy.5']
        #extrct interval
        inv=odf[odf['sum_occupancy']>1]
        #calculate interstitials' site type
        inv_s=np.zeros(5)
        for i in range(5):
            site_type=i+1
            inv_s[i]=len(inv[inv['Particle Type']==site_type])
        inv_s=inv_s/np.sum(inv_s)  
        inv_sites[n,:]=inv_s  
        #extrct vacancy
        v=odf[odf['sum_occupancy']<1]
        #calculate vacancies' site type
        v_s=np.zeros(5)
        for i in range(5):
            site_type=i+1
            v_s[i]=len(v[v['Particle Type']==site_type])
        v_s=v_s/np.sum(v_s) 
        v_sites[n,:]=v_s
        n+=1
    interval_sites=pd.DataFrame(inv_sites,columns = ['Ga1', 'Ga2','O1','O2','O3'])
    vacancy_sites=pd.DataFrame(v_sites,columns = ['Ga1', 'Ga2','O1','O2','O3'])
    pka_interval_sites.append(interval_sites)
    pka_vacancy_sites.append(vacancy_sites)
        
            
```

```python
element=['Ga1', 'Ga2','O1','O2','O3']
inv=dict()
v=dict()
for i in range(5):
    pka='PKA '+element[i]
    inv[pka]=pka_interval_sites[i].mean()
    v[pka]=pka_vacancy_sites[i].mean()
inv=pd.DataFrame(inv)
v=pd.DataFrame(v)
inv
```

```python
inv.plot(kind='bar', rot=0, alpha=1, figsize=(16,10),fontsize=20,capsize=5).grid(axis='y')
plt.xlabel('interstitials sites type',fontsize=30)
plt.ylabel('percentage',fontsize=30)
plt.rcParams["axes.linewidth"] = 3
```

```python
v.plot(kind='bar', rot=0, alpha=1, figsize=(16,10),fontsize=20,capsize=5).grid(axis='y')
plt.xlabel('Vacancies sites type',fontsize=30)
plt.ylabel('percentage',fontsize=30)
plt.rcParams["axes.linewidth"] = 3
```
