```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
os.chdir('/home/local/heruhe/Desktop/Ga2o3/cascade/5type/results') 
print(os.getcwd()) 

```

```python
fns=os.listdir()
print(fns)
for fn in fns:
    energy=fn.split('.')
    t,T,pe,ke,Et,p,v,festop= np.loadtxt(fn, unpack=True)
    plt.plot(t,T,label='PKA energy: '+energy[0])

plt.xlabel('Time (ps)',fontsize=30)
plt.ylabel('Temperature(K)',fontsize=30)
plt.rcParams["figure.figsize"] = (16,10)
plt.rcParams["axes.linewidth"] = 3
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
```
