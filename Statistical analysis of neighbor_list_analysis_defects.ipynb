```python
#Statistical analysis of neighbor_list_analysis defects
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
%matplotlib inline
import seaborn as sns
```

```python
path1='/home/heruhe/Desktop/Ga2o3/cascade/5type_overlapping/500ev/out'
os.chdir(path1)
```

```python
ndefects=np.zeros((6,5))
for f in range(1,250):
    ndf=pd.read_csv('frame{}/defects{}.csv'.format(f,f))
    for i in range(5):
        ndefects[f,i]=len(ndf[ndf['Particle Type']==i+1])
len(ndefects)
```

```python
#plot different type defects amount
def plot_defects(ndefects):
    elements=['Ga1', 'Ga2','O1','O2','O3']
    for i in range(len(ndefects)):
        plt.plot(elements,ndefects[i,:],'-o',label='Number of cascades{}'.format(i+1),linewidth=3)
    plt.rcParams["figure.figsize"] = (12,10)
    plt.xticks(fontsize=20,weight='bold')
    plt.yticks(fontsize=20,weight='bold')
    plt.legend(fontsize=20)
    plt.xlabel('Defects type',fontsize=20,weight='bold')
    plt.ylabel('Number of defects',fontsize=20,weight='bold')
    plt.rcParams["axes.linewidth"] = 3
    plt.show()
```

```python
plot_defects(ndefects)
```

```python
sdefects=ndefects.sum(axis=1)
plt.plot([i+1 for i in range(6)],sdefects,'-o',linewidth=3)
plt.rcParams["figure.figsize"] = (12,10)
plt.xticks(fontsize=20,weight='bold')
plt.yticks(fontsize=20,weight='bold')
plt.legend(fontsize=20)
plt.xlabel('Number of cascades',fontsize=20,weight='bold')
plt.ylabel('Total Number of defects',fontsize=20,weight='bold')
plt.rcParams["axes.linewidth"] = 3
plt.show()
```
