```python
def deduct_baseline(a,b,degree):
# Fit a polynomial to the data
    x=np.zeros(2)
    x[0]=a.iloc[0]
    x[1]=a.iloc[-1]
    y=np.zeros(2)
    y[0]=b.iloc[0]
    y[1]=b.iloc[-1]
    coeffs = np.polyfit(x, y, degree)
    baseline = np.polyval(coeffs, a)
    
# Subtract the baseline from the original signal
    corrected_signal = b - baseline

    return b 
```

```python
#alpha-amo structure analysis data
def structrue(Structure_type,path,pka):
    structrue=[]
    for i in pka:
        fn=path+'s.{}'.format(i)
        # Open the file in read mode
        with open(fn, "r") as file:
            # Read all lines from the file
            lines = file.readlines()

        # Extracting the structure counts data
        data_lines = [line.strip().split() for line in lines if not line.startswith("#")]
        df = pd.DataFrame(data_lines, columns=["Structure type", "Count"])
        structrue_row = df[df["Structure type"] == Structure_type]
        structrue.append(int(structrue_row["Count"].values[0]))
        
    return structrue
    
```
