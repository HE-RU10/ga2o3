```python
def ave_stresstensor(fn, bins):
    df=pd.read_csv(fn,sep=' ',skiprows=8)
    df.columns=[ 'id', 'type', 'x', 'y', 'z', 'c_ep', 'c_ek',
       'Coordination', 'ColorR', 'ColorG', 'ColorB', 'c_mypress1',
       'c_mypress2', 'c_mypress3', 'c_mypress4', 'c_mypress5', 'c_mypress6','ITEM:', 'ATOMS']
    df['binned_column'] = pd.cut(df['x'],bins=nbins)
    xl=df.groupby('binned_column')['x'].mean()
    xx = df.groupby('binned_column')['c_mypress1'].mean()
    plt.plot(xl,xx,label='xx')
    yy = df.groupby('binned_column')['c_mypress2'].mean()
    plt.plot(xl,yy,label='yy')
    zz = df.groupby('binned_column')['c_mypress3'].mean()
    plt.plot(xl,zz,label='zz')
    xy = df.groupby('binned_column')['c_mypress4'].mean()
    xz = df.groupby('binned_column')['c_mypress5'].mean()
    yz = df.groupby('binned_column')['c_mypress6'].mean()
    #plt.plot(xl,yz,label='yz')
    #plt.plot(xl,xy,label='xy')
    #plt.plot(xl,xz,label='xz')
    plt.legend()
```
