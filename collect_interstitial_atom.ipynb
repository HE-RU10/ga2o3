```python

#read defects sites original position
odf=pd.read_csv('Idefects_original_position1.csv')
#read interstitial atoms final position
ddf=pd.read_csv('Idefects_displaced_position1.csv')
ddf.drop(columns=ddf.columns[0], axis=1, inplace=True) #delete unneccessary columns
#delete original atoms in the defects sites, obtain interstitial atoms position info
I_atomsdf=ddf.drop(pd.Index(np.where(ddf['Particle Identifier'].isin(odf['Particle Identifier']))[0]))
I_atomsdf.reset_index(drop=True, inplace=True)
I_atomsdf
```

```python
#obtain defects atoms' neighbor list, combine the list into defects atoms dataframe
def calculate_neighbor(I_atomsdf,df):
    #I_atomsdf is defects atoms dataframe,including position & index& type info
    #df is all atoms dataframe,,including position & index& type info
    neighbors=np.zeros((len(I_atomsdf['Particle Identifier']),8))
    for i in range(len(I_atomsdf['Particle Identifier'])):
        index=I_atomsdf['Particle Identifier'][i]
        #obtain interstitial atom's positon
        x=I_atomsdf[I_atomsdf['Particle Identifier']==index].values[0][1]
        y=I_atomsdf[I_atomsdf['Particle Identifier']==index].values[0][2]
        z=I_atomsdf[I_atomsdf['Particle Identifier']==index].values[0][3]
        #regard the interstitial atom as center, choose a cube distrcition, extract all atoms info in the cube  
        cube_df = df[(np.abs(df.x-x)<3)& (np.abs(df.y-y)<3)&(np.abs(df.z-z)<3)]
        #In the cube, calculate each atom's distance with the center interstitial atom
        cube_df.loc[:,'distance']=np.sqrt((cube_df.x-x)**2+(cube_df.y-y)**2+(cube_df.z-z)**2)
        cube_df.sort_values(by=['distance'],inplace=True)
        #extract 8 nearest atoms' particle type info to obtain neighbor list
        neighbors[i,:]=cube_df['Particle Type'][1:9]
    I_atomsdf = pd.concat([I_atomsdf, pd.DataFrame(neighbors)], axis=1)
    return I_atomsdf
```

```python

for s in success_s:    
    #reading all atoms final position into df
    df=pd.read_csv('data.lastframe{}'.format(s), skiprows = 17, nrows=160000,sep=" ")

    df=df.reset_index()
    df.drop(columns=df.columns[-3:], axis=1, inplace=True)
    df.columns=['Particle Identifier','Particle Type','x','y','z']
    #read defects sites original position
    odf=pd.read_csv('Idefects_original_position{}.csv'.format(s))
    #read interstitial atoms final position
    ddf=pd.read_csv('Idefects_displaced_position{}.csv'.format(s))
    ddf.drop(columns=ddf.columns[0], axis=1, inplace=True) #delete unneccessary columns
    #delete original atoms in the defects sites, obtain interstitial atoms position info into I_atomsdf
    I_atomsdf=ddf.drop(pd.Index(np.where(ddf['Particle Identifier'].isin(odf['Particle Identifier']))[0]))
    I_atomsdf.reset_index(drop=True, inplace=True)
    new_Iatom=calculate_neighbor(I_atomsdf,df)
    #save new_Iatom df as csv file
    os.mkdir( "defects_analysis")
    directory = "defects_analysis/frame{}".format(s)
    path2 = os.path.join(path1, directory)
    os.mkdir(path2)
    os.chdir(path2)
    new_Iatom.to_csv('new_Iatoms{}.csv'.format(s)) 
    os.chdir(path1)
```
