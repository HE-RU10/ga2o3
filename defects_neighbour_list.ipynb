```python
#ndf=pd.read_csv('ndf_100.csv')
ndf=pd.read_csv('ndf8.csv')
ndf.drop(ndf.columns[0],axis=1,inplace=True)

ndf.iloc[:,5:]
```

```python
##Ga1 perfect crystal

nga1=ndf[ndf['Particle Type']==1] #all Ga1 df
nga1['nsum4']= nga1.apply(lambda row: row['n1'] + row['n2']+row['n3']+row['n4'], axis=1)
nga1['nsum6']= nga1.apply(lambda row: row['n1'] + row['n2']+row['n3']+row['n4']+row['n5']+row['n6'], axis=1)

npga1=nga1[(nga1['nsum4']==16)|(nga1['nsum6']==25)] #perfect Ga1 crystal
print(len(npga1))
```

```python
##Ga2 perfect crystal
nga2=ndf[ndf['Particle Type']==2] #all Ga2 df
nga2['nsum4']= nga2.apply(lambda row: row['n1'] + row['n2']+row['n3']+row['n4'], axis=1)
nga2['nsum6']= nga2.apply(lambda row: row['n1'] + row['n2']+row['n3']+row['n4']+row['n5']+row['n6'], axis=1)
npga2=nga2[(nga2['nsum4']==16)|(nga2['nsum6']==25)] #Ga2 perfect crystal df
print(len(npga2))
```

```python


ndga1=nga1[~nga1.isin(npga1)].dropna()#Ga1 defects
dropl=[]
for i in ndga1['Particle Identifier']:
    x=ndga1[ndga1['Particle Identifier']==i]['x'].values[0]
    z=ndga1[ndga1['Particle Identifier']==i]['z'].values[0]
    n=len(npga2[(abs(npga2['x']-x)<0.9)&(abs(npga2['z']-z)<0.9)])+len(npga1[(abs(npga1['x']-x)<0.9)&(abs(npga1['z']-z)<0.9)])  
    if n>2:
        print(i,n)
        dropl.append(int(i-1))
ndga1 = ndga1.drop(index=dropl)
print(len(ndga1))
ndga1.iloc[:,0:]
```

```python
#Ga2 defects

ndga2=nga2[~nga2.isin(npga2)].dropna()#Ga2 defects
dropl=[]
for i in ndga2['Particle Identifier']:
    x=ndga2[ndga2['Particle Identifier']==i]['x'].values[0]
    z=ndga2[ndga2['Particle Identifier']==i]['z'].values[0]
    n=len(npga2[(abs(npga2['x']-x)<0.9)&(abs(npga2['z']-z)<0.9)])+len(npga1[(abs(npga1['x']-x)<0.9)&(abs(npga1['z']-z)<0.9)])    
    if n>2:
        print(i,n)
        dropl.append(int(i-1))
ndga2 = ndga2.drop(index=dropl)
print(len(ndga2))

ndga2.iloc[:,0:]
```

```python
##O1 perfect crystal
no1=ndf[ndf['Particle Type']==3] #all O1
no1['nsum3']= no1.apply(lambda row: row['n1'] + row['n2']+row['n3'], axis=1)
no1['nsum4']= no1.apply(lambda row: row['n1'] + row['n2']+row['n3']+row['n4'], axis=1) 
npo1=no1[(no1['nsum3']==5)|(no1['nsum3']==4)|(no1['nsum4']==7)] #perfect O1 
print(len(npo1))
```

```python
##O2 perfect crystal
no2=ndf[ndf['Particle Type']==4]
no2['nsum3']= no2.apply(lambda row: row['n1'] + row['n2']+row['n3'], axis=1)
no2['nsum4']= no2.apply(lambda row: row['n1'] + row['n2']+row['n3']+row['n4'], axis=1) 
npo2=no2[(no2['nsum3']==5)|(no2['nsum3']==4)|(no2['nsum4']==7)] #perfect O2
print(len(npo2))
```

```python
##O3 perfect crystal
no3=ndf[ndf['Particle Type']==5]
no3['nsum3']= no3.apply(lambda row: row['n1'] + row['n2']+row['n3'], axis=1)
no3['nsum4']= no3.apply(lambda row: row['n1'] + row['n2']+row['n3']+row['n4'], axis=1) 
npo3=no3[(no3['nsum3']==5)|(no3['nsum3']==4)|(no3['nsum4']==7)] #perfect O3 
print(len(npo3))
```

```python
##O1 defects
ndo1=no1[~no1.isin(npo1)].dropna()#o1 defects
for i in ndo1['Particle Identifier']:
    x=ndo1[ndo1['Particle Identifier']==i]['x'].values[0]
    z=ndo1[ndo1['Particle Identifier']==i]['z'].values[0]
    n=len(npo1[(abs(npo1['x']-x)<0.4)&(abs(npo1['z']-z)<0.4)])+ len(npo2[(abs(npo2['x']-x)<0.4)&(abs(npo2['z']-z)<0.2)])+len(npo3[(abs(npo3['x']-x)<0.4)&(abs(npo3['z']-z)<0.4)])   
    if n>2:
        print(i,n)
        ndo1 = ndo1.drop(index=[int(i-1)])
print(len(ndo1))

ndo1
```

```python
##O2 defects
ndo2=no2[~no2.isin(npo2)].dropna()#o2 defects
for i in ndo2['Particle Identifier']:
    x=ndo2[ndo2['Particle Identifier']==i]['x'].values[0]
    z=ndo2[ndo2['Particle Identifier']==i]['z'].values[0]
    n=len(npo1[(abs(npo1['x']-x)<0.4)&(abs(npo1['z']-z)<0.4)])+ len(npo2[(abs(npo2['x']-x)<0.4)&(abs(npo2['z']-z)<0.2)])+len(npo3[(abs(npo3['x']-x)<0.4)&(abs(npo3['z']-z)<0.4)])
    if n>2:
        ndo2 = ndo2.drop(index=[int(i-1)])
print(len(ndo2))

ndo2
```

```python
##O3 defects
ndo3=no3[~no3.isin(npo3)].dropna()#o3 defects
for i in ndo3['Particle Identifier']:
    x=ndo3[ndo3['Particle Identifier']==i]['x'].values[0]
    z=ndo3[ndo3['Particle Identifier']==i]['z'].values[0]
    n=len(npo1[(abs(npo1['x']-x)<0.4)&(abs(npo1['z']-z)<0.4)])+ len(npo2[(abs(npo2['x']-x)<0.4)&(abs(npo2['z']-z)<0.2)])+len(npo3[(abs(npo3['x']-x)<0.4)&(abs(npo3['z']-z)<0.4)])
    if n>1:
        ndo3 = ndo3.drop(index=[int(i-1)])
print(len(ndo3))

ndo3
```
