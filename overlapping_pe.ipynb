```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
%matplotlib inline
import seaborn as sns
import re
```

```python
path='/home/local/heruhe/Desktop/Ga2o3/cascade/5type_overlapping/Ga/1500ev/log_file/'
os.chdir(path) 
pka=[]
pe=[]
for i in range(1,4301):
    pkaf=path+'lammps.stdout1_'+str(i)
    fout=str(i)+'.log'
    if  os.path.exists(pkaf):       
        extract_data="gawk "+ "'$1=="+ '"ec"'+" {print $2,$3,$5,$6,$7,$8,$9,$10}' " +pkaf+' > '+fout
        os.system(extract_data)
        pka.append(i)
        log= pd.read_csv(fout, sep=' ',header=None)
        log.columns=['step','t','T','pe','ke','Et','p','v']
        pe.append(list(log['pe'])[-1])
    else:
        continue
PE=pd.DataFrame()
PE['PKA']=pka
PE['pe']=np.array(pe)/160000
PE.to_csv('pe')
```

```python
path='/home/local/heruhe/Desktop/Ga2o3/cascade/5type_overlapping/500ev/out/log_file/'
os.chdir(path) 
pka=[]
pe=[]
for i in range(1,4301):
    pkaf=path+'lammps-anneal'+str(i)+'.log'
    fout=str(i)+'.log'
    if  os.path.exists(pkaf):       
        extract_data="gawk "+ "'$1=="+ '"ec"'+" {print $2,$3,$5,$6,$7,$8,$9,$10}' " +pkaf+' > '+fout
        os.system(extract_data)
        pka.append(i)
        log= pd.read_csv(fout, sep=' ',header=None)
        log.columns=['step','t','T','pe','ke','Et','p','v']
        pe.append(list(log['pe'])[-1])
    else:
        continue
PE=pd.DataFrame()
PE['PKA']=pka
PE['pe']=np.array(pe)/81920
PE.to_csv('anneal_pe')
```
