```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
values=[]
npartical=14400
for i in range(0,1000,10):
    fn='/home/heruhe/Downloads/jiahuijiahui-polikit-1847699d5efa/build/alpha/{}.txt'.format(i)
    
    with open(fn, 'r') as file:
            lines = file.readlines()
            if len(lines) >= 14:
                # Convert each number string to a float
                numbers = [float(num) for num in lines[13].split()]
                values.append(numbers)

# Convert the list of lists to a DataFrame
df = pd.DataFrame(values, columns=['CS', 'ES', 'FS'])
df['pka']=[i+1 for i in range(0,1000,10)]
plt.plot(df['pka'],df.CS/npartical,label='CS')
plt.plot(df['pka'],df.ES/npartical,label='ES')
plt.plot(df['pka'],df.FS/npartical,label='FS')
plt.scatter(1000,1800/11520)
plt.scatter(1000,2808/11520)
plt.scatter(1000,0/11520)
plt.xlabel('PKA numbers')
plt.ylabel('Counts')
plt.legend()
```
