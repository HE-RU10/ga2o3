```python
##delete original atoms in Iatoms file,just leaving displaced atoms in the file
f1 = open('Iatoms0.xyz', 'r') #read Iatoms file(ws atoms mode)
f2=open('Isites0.xyz', 'r') #read Isites file(ws sites mode)
Lines1 = f1.readlines()
Lines2 = f2.readlines()
#extract all atoms index info from Iatoms file
n1=int(Lines1[0]) #obtain atoms number in the file 
atoms=np.zeros(n1)
for i in range(n1):
    index=i+2
    atom=Lines1[index].split(' ')[0]
    #print(atom)
    atoms[i]=atom
#extract all sites index info from Iatoms file
n2=int(Lines2[0])  #obtain atoms number in the file 
sites=np.zeros(n2)
for i in range(n2):
    index=i+2
    site=Lines2[index].split(' ')[0]
    #print(atom)
    sites[i]=site
#obtain the indexs of original atoms in Iatoms file
indexs=[]
for i in sites:
    indexs.append(np.where(atoms==i)[0][0]+2)
#extract lines without original atoms info, and write all the displaced atoms info into a new xyz file- Idisplaced_atoms*.xyz
f3=open('Idisplaced_atoms0.xyz','w')
f3.write('{}\n'.format(n1-n2))
for i in range(1,n1+2):
    if i not in indexs:
        f3.write(Lines1[i])
```

```python
success_s=[]
for i in range(180):
    if os.path.isfile('./data.lastframe{}'.format(i)):
        success_s.append(i)
for s in success_s:
        ##delete original atoms in Iatoms file,just leaving displaced atoms in the file
    f1 = open('Iatoms{}.xyz'.format(s), 'r') #read Iatoms file(ws atoms mode)
    f2=open('Isites{}.xyz'.format(s), 'r') #read Isites file(ws sites mode)
    Lines1 = f1.readlines()
    Lines2 = f2.readlines()
    #extract all atoms index info from Iatoms file
    n1=int(Lines1[0]) #obtain atoms number in the file 
    atoms=np.zeros(n1)
    for i in range(n1):
        index=i+2
        atom=Lines1[index].split(' ')[0]
        #print(atom)
        atoms[i]=atom
    #extract all sites index info from Iatoms file
    n2=int(Lines2[0])  #obtain atoms number in the file 
    sites=np.zeros(n2)
    for i in range(n2):
        index=i+2
        site=Lines2[index].split(' ')[0]
        #print(atom)
        sites[i]=site
    #obtain the indexs of original atoms in Iatoms file
    indexs=[]
    for i in sites:
        try:
            indexs.append(np.where(atoms==i)[0][0]+2)
        except:
            continue
    #extract lines without original atoms info, and write all the displaced atoms info into a new xyz file- Idisplaced_atoms*.xyz
    f3=open('Idisplaced_atoms{}.xyz'.format(s),'w')
    f3.write('{}\n'.format(n1-n2))
    for i in range(1,n1+2):
        if i not in indexs:
            f3.write(Lines1[i])
```
