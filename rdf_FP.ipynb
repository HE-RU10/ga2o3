```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
%matplotlib inline
import seaborn as sns
from scipy import stats
```

```python
path='/home/heruhe/Desktop/Ga2o3/Frankpair/FP_alpha/rdf'
os.chdir(path)
f3='Ga-Gardf_alpha'
Gardf_alpha=pd.read_csv(f3, sep=' ',skiprows=1) 
f4='Ga-Gardf_gamma'
Gardf_gamma=pd.read_csv(f4, sep=' ',skiprows=1)  
f5='O-Ordf_alpha'
Ordf_alpha=pd.read_csv(f5, sep=' ',skiprows=1) 
f6='O-Ordf_gamma'
Ordf_gamma=pd.read_csv(f6, sep=' ',skiprows=1) 
l=[200,400,600] 
Pr_O_beta=np.zeros(3) 
Pr_O_gamma=np.zeros(3) 
Pr_Ga_beta=np.zeros(3) 
Pr_Ga_gamma=np.zeros(3) 
Pr_mix_beta=np.zeros(3) 
Pr_mix_gamma=np.zeros(3) 
for j in range(3): 
    i=l[j] 
    f='Ga-Gardf_'+'GaFP.'+str(i) 
    rdf=pd.read_csv(f, sep=' ',skiprows=1) 
    f1='Ga-Gardf_'+'OFP.'+str(i) 
    rdf2=pd.read_csv(f1, sep=' ',skiprows=1) 
    f2='Ga-Gardf_'+'mixFP.'+str(i) 
    rdf3=pd.read_csv(f2, sep=' ',skiprows=1) 
    annealf='Ga-Gardf'+'mixFPanneal.'+str(i) 
    #rdfanneal=pd.read_csv(annealf, sep=' ',skiprows=1)

    plt.plot(rdf.iloc[30:83,0],rdf.iloc[30:83,1]+j*5,label=f,linewidth=3)
    plt.plot(rdf2.iloc[30:83,0],rdf2.iloc[30:83,1]+j*5+2,label=f1,linewidth=3)
    plt.plot(rdf3.iloc[30:83,0],rdf3.iloc[30:83,1]+j*5+1,label=f2,linewidth=3)
#plt.plot(rdfanneal.iloc[40:,0],rdfanneal.iloc[40:,1]+j*3,'-+',label=annealf)
    Pr_Ga_beta[j]=stats.pearsonr(rdf.iloc[53:68,1], Gardf_beta.iloc[53:68,1])[0]
    Pr_Ga_gamma[j]=stats.pearsonr(rdf.iloc[53:68,1], Gardf_gamma.iloc[53:68,1])[0]
    Pr_O_beta[j]=stats.pearsonr(rdf2.iloc[53:68,1], Gardf_beta.iloc[53:68,1])[0]
    Pr_O_gamma[j]=stats.pearsonr(rdf2.iloc[53:68,1], Gardf_gamma.iloc[53:68,1])[0]
    Pr_mix_beta[j]=stats.pearsonr(rdf3.iloc[53:68,1], Gardf_beta.iloc[53:68,1])[0]
    Pr_mix_gamma[j]=stats.pearsonr(rdf3.iloc[53:68,1], Gardf_gamma.iloc[53:68,1])[0]
    print(Pr_Ga_beta) 
    print(Pr_Ga_gamma) 
    print(Pr_O_beta) 
    print(Pr_O_gamma) 
    print(Pr_mix_beta) 
    print(Pr_mix_gamma) 
plt.plot(Gardf_beta.iloc[30:83,0],Gardf_beta.iloc[30:83,1]-3,label=f3,linewidth=3) 
plt.plot(Gardf_gamma.iloc[30:83,0],Gardf_gamma.iloc[30:83,1]+j*7+1,label=f4,linewidth=3)

plt.xlabel('Pair distance,r($\AA$)',fontsize=30) 
plt.ylabel('PRDF,gpGa-Ga(r))',fontsize=30)
plt.rcParams["axes.linewidth"] = 3 
plt.rcParams["figure.figsize"] = (10,18) 
plt.xticks(fontsize=20) 
plt.yticks(fontsize=20) #plt.legend(fontsize=10) 
plt.yticks([]) 
plt.margins(x=0) 
plt.margins(y=0.02) 
plt.ylabel('PRDF, $g_{p}$Ga-Ga(r)', labelpad=20, fontsize=30)
plt.rcParams["font.family"] = "arial"
```
