```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
%matplotlib inline
import seaborn as sns
import scipy
```

```python
d=[3,6,9]
pGa=np.zeros(len(d))
for i in range (len(d)):
    path='/home/heruhe/Desktop/Ga2o3/Frankpair/Raw_data_hehuan/singleFP'+'/betaphase_GaFP_{}angstrom'.format(d[i])
    os.chdir(path) 
    df=pd.read_csv('interstitial.txt')
    count = (df.iloc[:,0] != 0).sum()
    pGa[i]=count/50
pGa
```

```python
dO=[3,9]
pO=np.zeros(len(dO))
for i in range (len(dO)):
    path='/home/heruhe/Desktop/Ga2o3/Frankpair/Raw_data_hehuan/singleFP'+'/betaphase_OFP_{}angstrom'.format(dO[i])
    os.chdir(path) 
    df=pd.read_csv('interstitial.txt')
    count = (df.iloc[:,0] != 0).sum()
    pO[i]=count/50
pO
```

```python
plt.plot(d,pGa*100,'-o',label='Ga FPs', linewidth=3)
plt.plot(dO,pO*100,'-o',label='O FPs', linewidth=3)
plt.rcParams["figure.figsize"] = (16,12)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=30)
plt.xlabel('Displacement distance ($\AA$)',fontsize=30)
plt.ylabel('Probability of creating a stable Frenkel pair (%)',fontsize=30)
plt.rcParams["axes.linewidth"] = 3
plt.rcParams["font.family"] = "arial"
```
