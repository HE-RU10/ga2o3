```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
path='/Users/ruhe/Desktop/ga2o3/FP/soapGap/FP/mix/log_file'
os.chdir(path) 
nFP=[i for i in range(1,200)]
ape_mix=np.zeros(199)
avol_mix=np.zeros(199)
for i in range(1,200):
    fn2=str(i)+'.log'
     # custom step dt time temp press v_ape v_ake v_avol etotal c_FP_distance
    log= pd.read_csv(fn2, sep=' ',header=None)
    log.columns=['Timestep','t','T','press','ape','ake','avol','etotal','FP_distance']
    ape_mix[i-1]=log['ape'][-1:]
    avol_mix[i-1]=log['avol'][-1:]
fig, ax1 = plt.subplots()
ax1.spines['right'].set_position(('axes', 1.0))
ax1.plot(nFP, ape_mix,color='tab:red')
ax1.set_ylabel('Potential energy (eV/atom)',color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')
ax1.set_xlabel('FP numbers')
ax2 = ax1.twinx()
ax2.plot(nFP, avol_mix,color='tab:green')
# Offset the last y-axis to the right
#ax2.spines['right'].set_position(('axes',1.2))
ax2.set_ylabel('Atomic volume $\mathrm{\AA^3/atom}$',color='tab:green')
ax2.tick_params(axis='y', labelcolor='tab:green')
```

```python


import pandas as pd
import io

# Read the data from the file
with open('/Users/ruhe/Desktop/ga2o3/FP/soapGap/FP/mix/displaced_atom', 'r') as file:
    data = file.read()


# Split data into blocks based on "displacement atom"
data_blocks = data.strip().split("displacement atom\n")

# Initialize an empty list to store DataFrames
dfs = []

# Loop through each block and create DataFrames
for block in data_blocks:
    # Skip empty blocks
    if not block:
        continue
    
    # Convert the block to a file-like object
    block_file = io.StringIO(block)
    # Read the data into a DataFrame and skip the first row (header)
    df = pd.read_csv(block_file, sep='\s+', skiprows=0)
    #df.columns=['times', 'ID', 'type',' distance','dis_x', 'dis_y', 'dis_z']
    dfs.append(df)
    #print(df)

# Concatenate all DataFrames into one
final_df = pd.concat(dfs, ignore_index=True)

# Print the final DataFrame
print(final_df)
i=1
print(final_df['dis_x'][i-1])
```
