```python

import numpy as np
import gzip
import glob
import subprocess
import time
from shutil import copy2
#test the function in xinjin's merge box script:merRelax_xinjin.py
def getCellList(path, character = '*.xyz'):
    """
    Get the list of cells and sorted according to the number along the z axis.
    
    path: path to the cell folder;

    character: the common character of the cells
    
    return:
        the list of the cell names sorted according to the number along the z axis.
    """
    
    file_name = []
    file_number = []
    
    path = path + character
    for i in glob.glob(path):
        file_name.append(i)
        tmp=i.split('/')[-1]
        tmp=tmp.split('.')[0]
        file_number.append(int(tmp)) 
    file_number, file_name = zip(* sorted(zip(file_number, file_name) ) )
        
    print("The list of files:")
    
    for i in range(len(file_name)):
        print(file_number[i], file_name[i])
        
    return(file_name)
cellList = getCellList('/home/heruhe/Desktop/Ga2o3/combine_cells/010_25boxes/boxes/', character = '*.xyz')

```

```python
import re
def readcells(inpath):
    """
    Read the cells.
    inpath: path to the cell;
    element: element name.
    """
    
    print("Start to read cells: %s\n" % inpath)
    
    coord = []
    eleSequence = []
    with open(inpath) as f:
        lines = [line.rstrip() for line in f]
        lineCount = 0
        for line in lines:
            lineCount += 1
            
            if lineCount == 1:
                totNum = int(line)
                
            elif lineCount == 2:
                if 'Lattice' in line and 'Origin' in line:
                    boxsize = re.findall(r'[-+]?\d*\.\d+|\d+', line)
                    boxsize = [float(num) for num in boxsize]

                    boxsize_x =boxsize[0]
                    boxsize_y =boxsize[4]
                    boxsize_z =boxsize[8]

                else:
                    raise KeyError("The format of cell has problem!")
            else:
                
                linesplit = line.split()
                element=int(linesplit[-1])
                if element>2:
                    eleSequence.append('O')
                elif element<3:
                    eleSequence.append('Ga')
                coord.append([])
                coord[-1].append(float(linesplit[0]))
                coord[-1].append(float(linesplit[1]))
                coord[-1].append(float(linesplit[2]))
                
    coord = np.array(coord)
    eleSequence = np.array(eleSequence)
    print(eleSequence)
    
    return(totNum, boxsize_x, boxsize_y, boxsize_z, coord, eleSequence)
dump1, tmp_x, tmp_y, tmp_z, dump2, dump3 = readcells(cellList[0])
```
